```
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
training_output=pd.read_excel('training_output.xlsx')
test_output=pd.read_excel('test_output.xlsx')
training_feature_matrix=pd.read_excel('training_feature_matrix.xlsx')
test_feature_matrix=pd.read_excel('test_feature_matrix.xlsx')
```

```
#data preprocessing
x=training_feature_matrix.to_numpy()
y=training_output.to_numpy()
x_bar=np.mean(x,axis=0)
x_std=np.std(x,axis=0)
x_new=(x-x_bar)/(x_std)
column_of_ones=np.ones(244).T
x_new2=np.insert(x_new, 0,column_of_ones, axis=1)
y_bar=np.mean(y,axis=0)
y_std=np.std(y,axis=0)
y_new=(y-y_bar)/(y_std)
```

```
#convert data to numpy arrays
x_test=test_feature_matrix.to_numpy()
y_test=test_output.to_numpy()
x_test_new=(x_test-x_bar)/(x_std)
ones_column=np.ones(103)
x_test_new2=np.insert(x_test_new,0,ones_column, axis=1)
y_test_new=(y_test-y_bar)/(y_std)
m,z=np.shape(x_new2)
```

```
#cost function
def cost_func(x,y,w):
  J=0
  for i in range(m):
    J=J+(0.5/m)*(np.dot(w,x[i].T)-y[i])**2
  return J
```

```
#gradient descent
m,z=np.shape(x_test_new2)
w=np.zeros((1,3))
W=np.zeros((m,3))
JJ=np.zeros((m,1))
def gradient_descent(X,Y,alpha):
  mse=0
  W[0,0]=w[0,0]
  W[0,1]=w[0,1]
  W[0,1]=w[0,1]
  for k in range(m):
    for i in range(m):
      for j in range(3):
        w[0,j]=w[0,j]-alpha*(np.dot(w,X[i].T)-Y[i])*X[i,j]
    W[k]=w
    JJ[k]=cost_func(X,Y,w)
  for i in range(m):
    diff_square=(np.dot(w,X[i].T)-Y)**2
    mse=mse+diff_square
  mse=mse/m
  plt.plot(JJ)
  plt.xlabel('Iterations',)
  plt.ylabel('Cost')
  a=W[:,0]
  b=W[:,1]
  c=W[:,2]
  p=JJ.reshape((103,))
  fig = plt.figure()
  ax = fig.gca(projection='3d')
  ax.plot(a,b,p)
  ax.set_xlabel('W1')
  ax.set_ylabel('W2')
  ax.set_zlabel('Cost')
  return (mse[m-1])
  

```

```
#function to test the model on test data
m,z=np.shape(x_test)
def test_model(x,y):
  predicted_values=np.zeros((m,1))
  mse=0
  for i in range(m):
    predicted_values[i]= W[m-1,0]*x[i,0]+W[m-1,1]*x[i,1]+W[m-1,2]*x[i,2]
  for i in range(m):
    diff_square=(predicted_values[i]-y[i])**2
    mse=mse+diff_square
  mse=mse/m
  print("MSE=")
  return mse[0]
  
```
